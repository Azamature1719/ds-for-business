```python
import pandas as pd
```

```python
df = pd.read_csv('/home/anvar/CC GENERAL.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.count()
```

```python
columns = df.columns

X = df[columns[1:]]
```

```python
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=8)
```

```python
kmeans.fit(X)
```

```python
kmeans.predict(X)
```

```python
df['clusters_kmeans_8'] = kmeans.predict(X)
```

```python
df['clusters_kmeans_8'].value_counts()
```

```python
df.query('clusters_kmeans_8 == 2')
```

# Try different K

```python
from tqdm import tqdm
```

```python
inertia = []
clusters = range(2, 40)

for k in tqdm(clusters):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    df[f'clusters_kmeans_{k}'] = kmeans.predict(X)
    inertia.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(clusters, inertia)
```

```python
def detect_outliers_simple(x):
    x = x.values
    q3 = np.quantile(x, 0.75)
    q2 = np.quantile(x, 0.25)
    iqr = q3-q2
    lower_outliers = (q2-1.5*iqr) > x
    upper_outliers = (q3+1.5*iqr) < x
    return lower_outliers + upper_outliers
```

```python
columns = df.columns
```

```python
cols_outliers = dict()
outliers = [False] * df.shape[0]

for col in columns[1:]:
        cols_outliers[col] = detect_outliers_simple(df[col])
        outliers += cols_outliers[col]
        print(col, np.round(cols_outliers[col].sum() / df.shape[0] * 100), '%')
```

```python
X_drop_outliers = X[~outliers]
```

```python
inertia = []
clusters = range(2, 40)

for k in tqdm(clusters):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_drop_outliers)
    X_drop_outliers[f'clusters_kmeans_{k}'] = kmeans.predict(X_drop_outliers)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(clusters, inertia)
```

```python
from sklearn.cluster import AgglomerativeClustering

ac = AgglomerativeClustering(n_clusters=10,
                             affinity='euclidean',
                             linkage='complete')

prediction = ac.fit_predict(X[~outliers])
print('Cluster labels: %s\n' % prediction)
```

```python
np.unique(prediction, return_counts=True)
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
```

```python
clusters = linkage(X[~outliers][::10], metric='euclidean', method='complete')
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,8))
dendr = dendrogram(clusters, ax=ax)

plt.ylabel('Euclidean Distance');
```
